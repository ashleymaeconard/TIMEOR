```python
import pandas as pd
import glob, os
import seaborn as sns
sns.set()
import matplotlib.pyplot as plt
```

```python
truth = pd.read_csv("/nbu/compbio/aconard/simulation/sim_all_results/timeor/results/analysis/ground_truth_results/temporal_relations_40/TF_tel_rel.csv").drop(['cluster'], axis=1)
```

```python
def cal_prec_recall(tr, si):
    # Get number tp, fp, fn
    df_m = pd.merge(tr, si,on=["regulator", "regulated", "regulation_type", "edge_type"], 
                    how='right', indicator="Exist")
    
    num_common_rows = (df_m['Exist'].value_counts().filter(items=['both']))[0]
    num_sim_rows = (df_m['Exist'].value_counts().filter(items=['right_only']))[0]
    tp = num_common_rows
    fn = (len(truth)- num_common_rows)
    fp = num_sim_rows
    
    print(df_m)
    
    prec = (tp / (tp + fp))
    rec = (tp / (tp + fn))
    
    return(prec, rec)
```

```python
truth
```

```python
prec = []
rec = []
perconc = []

for tf_rel in glob.glob('/nbu/compbio/aconard/simulation/sim_all_results/timeor/results/analysis/sim-pt05_results/te*/TF_tel_rel.csv', recursive=True):
    perc_concor = (os.path.basename(os.path.dirname(tf_rel)).split("_")[-1])
    sim = pd.read_csv(tf_rel).drop(['cluster'], axis=1)
    p,r = cal_prec_recall(truth, sim)
    
    print(perc_concor)
    print("precision", p)
    print("recall ", r)
    
    prec.append(p)
    rec.append(r)
    perconc.append(int(perc_concor))
```

```python
print(prec)
print(rec)
print(perconc)
```

```python
# Plot precision and recall

print(len(prec), len(rec), len(perconc))

df = pd.DataFrame(list(zip(prec, rec, perconc)), 
               columns =['precision', 'recall', 'percent_concordance']) 

df.sort_values(by=['percent_concordance'], inplace=True)
print(df)
fig, ax = plt.subplots(figsize=(10,10))
#sns.set_palette(reversed(sns.color_palette("Blues_d", n_plots)), n_plots)
plt.plot(df['percent_concordance'], df['precision'], color='blue')
#ax = sns.lineplot(x="recall", y="precision", data=df, ci=None)

ax = sns.scatterplot(x="percent_concordance", y="precision", data=df,color="black",legend=False)

plt.ylim(0,1.05)
#plt.xlim(0,1)

plt.savefig("/data/compbio/aconard/TIMEOR/simulation_hsa/precision_simulation_TIMEOR.pdf", dpi=300)

# def label_point(x, y, val, ax):
#     a = pd.concat({'x': x, 'y': y, 'val': val}, axis=1)
#     for i, point in a.iterrows():
#         ax.text(point['x']+0.01, point['y']-0.01, str(point['val']))

# label_point(df.recall, df.percent_concordance, df.percent_concordance, plt.gca())

# for line in range(0,df.shape[0]):
#      ax.text(df.precision[line], df.recall[line], df.percent_concordance[line], horizontalalignment='center', size='small', color='black', weight='semibold')
```
